### To see how many matches have been played in any particular city

```python
def matches(city):
    mask=df['city']==city
    return df[mask].shape[0]
cityname=input()
matches(cityname)
```

### To see how many matches have been won by a particular or more runs

```python
def winbyruns(runs):
    mask=df['win_by_runs']>runs
    return df[mask].shape[0]
runs=int(input())
winbyruns(runs)
```

### To see who has become the most man of the match in a particular city

```python
def manofthematch(city):
    mask=df['city']==city
    newdf=df[mask]
    mask1=newdf['player_of_match'].value_counts()
    return mask1
manofthematch(input().title())
```

### Drop Duplicates

It has three parameters:
1. subset (to select specific column)
2. keep   (which duplicate to keep first one or last one)
3. inplace( as default the value is false by making it true it will make changes in main dataset)

```python
df.drop_duplicates(subset=['city','season'],keep='first')
#droping from both column
```

### To see in which season who was the champion

basically champion will be the same team who win the last match that season

```python
x=df.drop_duplicates(subset=['season'],keep='last')
print(x[['season','winner']].sort_values('season'))
```

### Orange Cap Holders Season wise

```python
total=df.merge(delivery,left_on='id',right_on='match_id')
x=total.groupby(['season','batsman']).sum()["batsman_runs"]
x=x.reset_index()#covert series ito data frame
x=x.sort_values('batsman_runs',ascending=False).drop_duplicates('season')
x=x.sort_values('season')
print(x)
```

### Pivot Table
